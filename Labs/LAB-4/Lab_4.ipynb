```python
class base:
    
    def __init__(self,level):
        self.level=level
    



class Card:
    suits = ['Hearts','Spades','Clubs','Diamonds']
    values = [2,3,4,5,6,7,8,9,10,'Jack','Queen','King','Ace']
    
    def __init__(self,suit,value):  
        base.__init(self)
        self.__suit =suit
        self.__value = value
    
    def value(self):
        return self.__value
    def suit(self):
        return self.__suit
    def face_value(self):
        if self.__value == 'Ace':
            return 1
        elif self.__value == 'Jack' or self.__value == 'Queen' or self.__value == 'King':
            return 10
        else:
            return self.__value
        
    def shuffle_card(self):
        return self.__suit == "Shuffledcard"
    
    def __str__(self):
        if self.shuffledcard():
            return "Shuffledcard"
        else:
            return str(self.__value) + " of " + self.__suit
    
    
class Deck:
    def __init__(self,n_decks):
        self.__n_decks = n_decks
        self.__cards = []
        
        #for i in range(self.__n_decks):    
    def make_deck(self):
        values = [2,3,4,5,6,7,8,9,10,'Jack','Queen','King','Ace']
        suits = ['Hearts','Spades','Clubs',"Diamonds"]
        deck = []
    
        for i in values:
            for j in suits:
                deck.append((i,j))
        return deck 
    def shuffle(self):
        random.shuffle(self.__cards)
        
    def deal_cards(self):
        if len(self.__cards)>0:
            return self.__cards.pop()
        else:
            for i in range(self.__n_decks):
                self.__cards.extended(self.__make_deck())
            self.shuffle()
            return self.__cards.pop()
        
```

```python
def calc_hand_sum(hand):
    card_values = list(map(lambda card: card.face_value(),hand))
    
    hand_sum= sum(card_values)
    
    One_ace=hand_sum
    Eleven_ace=hand_sum+10
    
    if Eleven_ace<=21:
        return Eleven_ace
    else:
        return One_ace
```
